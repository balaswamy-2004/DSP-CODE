```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# IMPORT DATA
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')

# DESCRIBE DATA
print(df.head())

# DATA VISUALIZATION
print(df.info())

# Check for duplicate CustomerId
print("Number of duplicated CustomerId:", df.duplicated('CustomerId').sum())

# Set CustomerId as index
df = df.set_index('CustomerId')
print(df.info())

# DATA PREPROCESSING
print("Geography value counts:\n", df['Geography'].value_counts())

# Replace Geography values (corrected case sensitivity)
df.replace({'Geography': {'France': 2, 'Germany': 1, 'Spain': 0}}, inplace=True)

# Verify the replacement
print("Geography value counts after replacement:\n", df['Geography'].value_counts())

# TARGET VARIABLE Y, FEATURE VARIABLE X
print("Columns in dataframe:\n", df.columns)

# Print the first few rows to see the actual column names
print(df.head())

# Identify the correct column name for the target variable
# It appears that the target column is 'Exited' as per the data link, correct the name if necessary
target_column = 'Exited'  # Change this if the actual target column name is different

# Check if the target column exists in the DataFrame
if target_column in df.columns:
    # Drop Surname and target columns to create feature matrix X
    X = df.drop(['Surname', target_column], axis=1)

    # Target variable
    y = df[target_column]

    # Verify shapes of X and y
    print("Shape of X:", X.shape)
    print("Shape of y:", y.shape)

    # TRAIN TEST SPLIT
    from sklearn.model_selection import train_test_split

    # Corrected the assignment
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)

    # Verify shapes of train and test splits
    print("Shape of X_train:", X_train.shape)
    print("Shape of X_test:", X_test.shape)
    print("Shape of y_train:", y_train.shape)
    print("Shape of y_test:", y_test.shape)
else:
    print(f"Error: Column '{target_column}' not found in the DataFrame. Please check the column name.")
```
